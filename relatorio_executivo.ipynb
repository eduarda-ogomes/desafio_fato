# Relatório Executivo — Desafio Técnico (Estágio em Dados)

## 1. Contexto

Este projeto foi desenvolvido como parte de um desafio técnico para vaga de estágio em dados.  
O objetivo principal foi estruturar um fluxo completo de dados — desde a validação inicial até a geração de insights estratégicos — demonstrando organização, raciocínio analítico e capacidade de apoiar decisões de negócio.

Os dados simulam informações extraídas de um CRM e foram inicialmente disponibilizados em múltiplas abas de Excel (Vendas, Produtos, Clientes e Metas).

A partir disso, foi construído um pipeline simples, consistente e reprodutível utilizando Python, Pandas e SQLite.

---

## 2. Arquitetura do Projeto

O projeto foi organizado em três camadas bem definidas:

### 1. Tratamento e Certificação dos Dados

- Leitura única do Excel.
- Padronização estrutural das colunas.
- Conversão e validação de datas.
- Verificação de inconsistências.
- Correção de erro estrutural na receita.
- Persistência dos dados tratados em SQLite (fonte oficial).

O Excel deixa de existir após essa etapa, garantindo centralização e consistência.

---

### 2. Camada Analítica

- Leitura exclusiva do SQLite.
- Cálculo de receita, lucro e margem.
- Consolidação mensal dos indicadores.
- Análise de Meta vs Realizado.
- Estruturação de tabelas prontas para visualização.

Essa separação evita retrabalho no BI e garante que os números já estejam validados antes da visualização.

---

### 3. Dashboard Executivo (Metabase)

O dashboard foi construído para responder perguntas estratégicas, contendo:

- Receita Total
- Lucro Total
- Margem Média
- Atingimento de Meta
- Evolução mensal de Receita e Lucro
- Meta vs Realizado
- Receita por Canal
- Concentração por Produto

---

## 3. Correção Crítica de Dados

Durante a validação, foi identificado que o `valor_unitario` da aba Vendas não era confiável.  
O preço oficial estava na tabela Produtos.

### Decisão aplicada:
Recalcular toda a receita com base no preço oficial da tabela Produtos.

Essa decisão foi essencial para garantir consistência e confiabilidade nas análises posteriores.

---

## 4. Resultados Financeiros

### Indicadores do Período

- **Receita Total:** ~R$ 472 mil  
- **Lucro Total:** ~R$ 185 mil  
- **Margem Média:** ~39,8%

A margem média indica operação com rentabilidade sólida ao longo do período analisado.

Receita e lucro crescem de forma proporcional, o que demonstra crescimento sustentável — não apenas aumento de volume.

---

## 5. Análise de Crescimento e Metas

A meta disponível é mensal (meta de faturamento).

### Resultado:

- Meta atingida em **5 de 6 meses**
- Fevereiro foi o único mês abaixo da meta
- Crescimento consistente a partir de março

O atingimento elevado pode indicar:
- Meta conservadora
- Ou aceleração comercial no período

Em ambos os casos, a empresa apresenta desempenho positivo.

---

## 6. Análise por Canal

Os canais de venda apresentam distribuição relativamente equilibrada entre:

- Marketplace
- WhatsApp
- Site

Não há dependência extrema de um único canal, o que reduz risco operacional.

---

## 7. Concentração por Produto

O produto com maior faturamento representa aproximadamente **45% da receita total**.

Isso indica concentração relevante.

Embora o desempenho seja positivo, existe risco estratégico caso haja queda de demanda ou problema operacional nesse produto.

Recomenda-se acompanhamento contínuo da diversificação do portfólio.

---

## 8. Limitações do Dataset

- Ausência de identificador único de venda.
- Período analisado limitado (6 meses).
- Ausência de histórico mais longo para análise de sazonalidade.

Essas limitações foram consideradas ao longo da análise para evitar conclusões indevidas.

---

## 9. Conclusão

Os dados indicam:

- Crescimento consistente a partir do segundo trimestre.
- Receita e lucro evoluindo de forma proporcional.
- Margem estável.
- Metas amplamente atingidas.
- Concentração relevante em um produto específico.

O projeto demonstra:

- Organização em camadas.
- Separação clara entre tratamento, análise e visualização.
- Correção consciente de inconsistências.
- Construção de indicadores orientados à decisão.
- Capacidade de transformar dados brutos em informação estratégica.

A proposta foi desenvolver um fluxo simples, coerente e tecnicamente consistente, adequado ao nível de estágio, mas estruturado de forma profissional.
